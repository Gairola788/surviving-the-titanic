```python
import pandas as pd
import pickle

# Load cleaned / processed test data
test_final = pd.read_csv("test_cleaned.csv")

# ðŸ”¥ Fix: remove Survived if present in processed test file
if "Survived" in test_final.columns:
    test_final = test_final.drop(columns=["Survived"])

# Load original test.csv to extract PassengerId
test = pd.read_csv(r"C:\Users\Gairo\Downloads\titanic\test.csv")

# Load saved model
model = pickle.load(open("titanic_model.pkl", "rb"))

# Predict
predictions = model.predict(test_final)

# Create submission file
submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": predictions
})

submission.to_csv("submission.csv", index=False)

print("submission.csv generated successfully!")
```
